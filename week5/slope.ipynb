```python
import matplotlib.pyplot as plt

# my coordinates
x1 = 1
y1 = 2
x2 = 4
y2 = 5

# Develop the equation y = mx + b
m = (y2 - y1) / (x2 - x1) # my slope
b = y1 - m*x1 # my y-intercept
print("y = ", m, "x + ", b)

# For the graph
xmin = -10
xmax = 10
ymin = -10
ymax = 10

# For the line on the graph
y3 = m*xmin + b # my y3, I need a min value for the plot
y4 = m*xmax + b # my y4, I need a max value for the plot

# Basic setup for the graph
fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax],[0,0],'b') # blue x axis
plt.plot([0,0],[ymin,ymax], 'b') # blue y axis

# Plot the linear function as a red line
plt.plot([xmin,xmax],[y3,y4],'r') # here i'm using the coordinates for the min (-10, -9)

plt.show()
```
