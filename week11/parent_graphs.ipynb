Graphs with no slope and no x value<br>
y = 5

```python
import matplotlib.pyplot as plt
import numpy as np

xmin = -10
xmax = 10
ymin = - 10
ymax = 10
points = 4*(xmax-xmin)
x = np.linspace(xmin,xmax,points) 


fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax],[0,0],'-b') # blue x axis
plt.plot([0,0],[ymin,ymax], 'b') # blue y axis

plt.xticks(np.arange(xmin, xmax+1, 2))
plt.yticks(np.arange(ymin, ymax+1, 2))
plt.plot([xmin,xmax],[5,5],'r')

#ax.set_ylim(2, 4) # defining limits 
plt.show()
```

Cubic graphs <br>
y = x3

```python
import matplotlib.pyplot as plt
import numpy as np

xmin = -10
xmax = 10
ymin = -10
ymax = 10
points = 10 * (xmax - xmin)
x = np.linspace(xmin, xmax, points)

# Criar uma figura com 2 subplots em uma linha
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Primeiro subplot
axs[0].axis([xmin, xmax, ymin, ymax])
axs[0].plot([xmin, xmax], [0, 0], 'b')  # eixo x
axs[0].plot([0, 0], [ymin, ymax], 'b')  # eixo y
axs[0].set_xticks(np.arange(xmin, xmax, 1))
axs[0].set_yticks(np.arange(ymin, ymax, 1))
axs[0].plot(x, x**3)
axs[0].set_title('y = x')

# second subplot
axs[1].axis([xmin, xmax, ymin, ymax])
axs[1].plot([xmin, xmax], [0, 0], 'b')  #  x
axs[1].plot([0, 0], [ymin, ymax], 'b')  #  y
axs[1].set_xticks(np.arange(xmin, xmax, 1))
axs[1].set_yticks(np.arange(ymin, ymax, 1))
axs[1].plot(x, x**3 - 2*x**2 - 3*x + 4)
axs[1].set_title('y = x - 2x - 3x + 4')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

xmin = -10
xmax = 10
ymin = - 10
ymax = 10
points = 10*(xmax-xmin)
x = np.linspace(xmin,xmax,points) 

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax],[0,0],'b') # blue x axis
plt.plot([0,0],[ymin,ymax], 'b') # blue y axis

plt.xticks(np.arange(xmin,xmax,1))
plt.yticks(np.arange(ymin,ymax,1))
plt.plot(x, np.abs(x), 'r') 
plt.title('y = |x|')
plt.show()
```
