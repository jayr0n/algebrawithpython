# Exercises

exercises from Complete the problems on pages 304, 308, and 321 from ["Algebra and Trigonometry".](https://assets.openstax.org/oscms-prodcms/media/documents/Algebra-and-Trigonometry-2e-WEB.pdf?_gl=1*pbwdm8*_ga*MjA5NTQzMzA5NS4xNjk5NzIzMzQ0*_ga_T746F8B0QC*MTY5OTcyMzM0My4xLjEuMTY5OTcyMzYyOS40MC4wLjA.)

```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import symbols

def plotting_exercise(eq, xmin=-10, xmax=10, ymin=-10, ymax=10):
    points = 10 * (xmax - xmin)
    
    x_sym = symbols('x')  
    y = np.vectorize(lambda x: eq.subs(x_sym, x))(np.linspace(xmin, xmax, points))

    fig, ax = plt.subplots()
    plt.axis([xmin, xmax, ymin, ymax])  
    plt.plot([xmin, xmax], [0, 0], 'b')  
    plt.plot([0, 0], [ymin, ymax], 'b')  

    ax.set_xlabel("Valores de x")
    ax.set_ylabel("Valores de y")
    ax.set_title("Algum Gráfico")
    ax.grid(True)

    ax.set_xticks(np.arange(xmin, xmax, 1)) 
    ax.set_yticks(np.arange(ymin, ymax, 1)) 

    plt.plot(np.linspace(xmin, xmax, points), y, label=f'y = {eq}')
    plt.legend()
    plt.show()
```

## y = |x| - 2

```python
x = symbols('x')  # Defina x como um símbolo
eq = abs(x) - 2
plotting_exercise(eq)
```

## y = -|x|

```python
x = symbols('x')  # Defina x como um símbolo
eq = - abs(x)
plotting_exercise(eq)
```

## y = -|x| - 2

```python
x = symbols('x')  # Defina x como um símbolo
eq = - abs(x) - 2
plotting_exercise(eq)
```

## f(x) = -|x + 4| - 3

```python
x = symbols('x')  # Defina x como um símbolo
eq = - abs(x+4) - 3
plotting_exercise(eq)
```

## f(x) = 1/2 * |x + 4| - 3

```python
x = symbols('x')  # Defina x como um símbolo
eq = 1/2 * abs(x+4) - 3
plotting_exercise(eq)
```

<h1> Inverse function </h1>

<h2>$f(x)={x^{2}}$</h2>
<h2>$f(x)^{-1}={√x}$</h2>

```python
import matplotlib.pyplot as plt
import numpy as np
import math

# For the graph
xmin = -10
xmax = 10
ymin = -10
ymax = 10

# my x and limits
x1 = 0
x2 = 10
x = np.linspace(x1, x2, 100 * (x2 - x1))

# my y
y = x**2

# my inverse y
y2 = np.sqrt(x)

# my y = x
yx = np.linspace(xmin, xmax, 10 * (xmax - xmin))

# Setting my graph
fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax],[0,0],'b') # blue x axis
plt.plot([0,0],[ymin,ymax], 'b') # blue y axis

# Add details to the graph
ax.set_xlabel("time")
ax.set_ylabel("profit")
ax.grid(True)

ax.set_xticks(np.arange(xmin, xmax, 1)) # setting the "points" in x
ax.set_yticks(np.arange(ymin, ymax, 1)) # setting the "points" in y

ax.set_ylim(-2, 4) # setting the limits in y like a zoom
ax.set_xlim(-2, 4) # setting the limits in x like a zoom

# plotting graph
plt.plot(x, y,'r')
plt.plot(x, y2,'c')
plt.plot(yx, yx, '--', color='orange')
plt.show()
```
